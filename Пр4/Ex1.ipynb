```python
import numpy as np

улица = [80, 98, 75, 91, 78]
гараж = [100, 82, 105, 89, 102]

correlation_matrix = np.corrcoef(улица, гараж)


correlation = correlation_matrix[0, 1]


print("Коэффициент корреляции между улицей и гаражом:", correlation)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
plt.grid(True)
plt.title('Диаграмма рассеяния между Улицей и Гаражом', fontsize = 15)
plt.xlabel('улица')
plt.ylabel('гараж')
plt.scatter(улица, гараж, marker = 'o', color = 'crimson')
```
