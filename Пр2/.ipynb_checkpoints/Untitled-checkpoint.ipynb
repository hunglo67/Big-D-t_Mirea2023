```python
import pandas as pd
DATA_PATH = '/Users/hoang/Desktop/Big-Data_Mirea2023/Пр2/insurance.csv'
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
df.hist(color = 'red', edgecolor = 'black', bins = 30)
plt.show()
```

```python
print("Индекс массы тела(bmi):")
print('Среднее: ', df["bmi"].mean())
print('Мода: ', df['bmi'].mode()[0])
print('Медиана: ', df['bmi'].median())
print('Размах: ', df['bmi'].max()-df['bmi'].mean())
print('Стандартное отклонение: ', df['bmi'].std())
print("-------------------------------")
print("Расход: ")
print('Среднее: ', df['charges'].mean())
print('Мода: ', df['charges'].mode()[0])
print('Медиана: ', df['charges'].median())
print('Размах: ', df['charges'].max() - df['charges'].mean())
print('Стандартное отклонение: ', df['charges'].std())
```

```python
import matplotlib.pyplot as plt

bmi_mode = df['bmi'].mode()[0]  
bmi_median = df['bmi'].median()

plt.figure(figsize=(12, 12))
plt.hist(df['bmi'], bins=20, edgecolor='black', color='blue', alpha=0.7, label='BMI')
plt.axvline(x=bmi_mean, color='red', label='Среднее')
plt.axvline(x=bmi_mode, color='green', label='Мода')
plt.axvline(x=bmi_median, color='yellow', label='Медиана')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

charges_mean = df['charges'].mean()
charges_mode = df['charges'].mode()[0]  # Mode may have multiple values, so we take the first one
charges_median = df['charges'].median()

plt.figure(figsize=(12, 12))

plt.hist(df['charges'], bins=20, edgecolor='black', color='blue', alpha=0.7, label='Charges')

plt.axvline(x=charges_mean, color='red', label='Среднее')
plt.axvline(x=charges_mode, color='green', label='Мода')
plt.axvline(x=charges_median, color='yellow', label='Медиана')

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd


numeric_features = df.select_dtypes(include=['float64', 'int64']).columns

for feature in numeric_features:
    plt.figure(figsize=(8, 6))
    df.boxplot(column=feature)
    plt.title(f"Box Plot для {feature}")
    plt.show()
```
