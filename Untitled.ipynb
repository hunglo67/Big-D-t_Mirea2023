```python
import pandas as pd
data = pd.read_csv('C:/Users/hoang/Desktop/Big-Data_Mirea2023/Пр4/insurance.csv')
data
```

```python
missing_values = data.isnull().sum()
print(missing_values)

```

```python
unique_regions = data['region'].unique()
print("\nУникальные регионы:")
for region in unique_regions:
    print(region)
```

```python
from scipy.stats import f_oneway

regions = data['region'].unique()
group = [data['bmi'][data['region'] == region] for region in regions]

f_statistic, p_value = f_oneway(*group)

print("Результаты однофакторного ANOVA теста:")
print(f"F-статистика: {f_statistic}")
print(f"P-значение: {p_value}")



```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

region_bmi = data[['bmi', 'region']].copy()
model = ols('bmi ~ C(region)', data=data).fit()

anova_table = anova_lm(model)

print("Результаты однофакторного ANOVA теста:")
print(anova_table)



```

```python
from scipy import stats

region_pair = []

for i in range(len(regions)):
    for j in range(i + 1, len(regions)):
        region_pair.append((regions[i], regions[j]))
region_pair 

```

```python
groups = data.groupby('region').groups
      
for i, j in region_pair:
    print(i+ " - " +j)
    print(stats.ttest_ind(region_bmi['bmi'][groups[i]],region_bmi['bmi'][groups[j]] ))
        
        
      
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(endog=data['bmi'], groups = data['region'], alpha = 0.05)
tukey.plot_simultaneous()
tukey.summary()
```

```python
model = ols('bmi ~ C(sex) + C(region) + C(sex):C(region)', data = data).fit()
sm.stats.anova_lm(model, type = 2)
```
